```python
import pandas as pd
```

```python
dataset1= pd.read_csv("finalheart.csv")
```

```python
dataset2= pd.read_csv("heart.csv")
```

```python

dataset1.head()

dataset2.head()
```

```python
exercise_mapping = {'N': 0, 'Y': 1}
dataset1['ExerciseAngina'] = dataset1['ExerciseAngina'].map(exercise_mapping)
slope_mapping = {'Up': 0, 'Flat': 1, '2': 2}
dataset1['ST_Slope'] = dataset1['ST_Slope'].map(slope_mapping)
sex_mapping = {'M': 0, 'F': 1}  
dataset1['Sex'] = dataset1['Sex'].map(sex_mapping)
```

```python
display(dataset1.head())
dataset1.drop(columns=["ChestPainType","RestingECG"],inplace=True)
display(dataset1.head())
```

```python
display(dataset2.head())
dataset2.drop(columns=["cp","ca","thal","restecg"],inplace=True)
display(dataset2.head())
```

```python
dataset2.rename(columns={'age':'Age','sex':'Sex','trestbps':'RestingBP','chol':'Cholesterol','fbs':'FastingBS','thalach':'MaxHR','exang':'ExerciseAngina','oldpeak':'Oldpeak','slope':'ST_Slope','target':'HeartDisease'}, inplace= True)
```

```python
display(dataset1.head())
display(dataset2)
```

```python
display(dataset1.shape,dataset2.shape)
```

```python
merged_dataset= pd.concat([dataset1,dataset2])
merged_dataset.shape
```

```python
merged_dataset.drop(columns=['ST_Slope'], inplace=True)
merged_dataset.head()
```

```python
merged_dataset.to_csv('merged_dataset2.csv', index=False)
```
